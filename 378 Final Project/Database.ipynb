```python
import sqlite3
import random
from datetime import datetime, timedelta
```

```python
# Establish a connection to the database or create it if it doesn't exist
conn = sqlite3.connect('library.db')

# Create a cursor object to execute SQL commands
cur = conn.cursor()

# Create table for Books
cur.execute('''
CREATE TABLE IF NOT EXISTS Books (
    Book_ID INTEGER PRIMARY KEY,
    Title TEXT NOT NULL,
    Author TEXT,
    ISBN TEXT,
    Publication_Year INTEGER,
    Category TEXT,
    Availability TEXT CHECK (Availability IN ('Yes', 'No')) NOT NULL
)
''')

# Create table for Students
cur.execute('''
CREATE TABLE IF NOT EXISTS Students (
    Student_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Email TEXT,
    Department TEXT
)
''')

# Create table for Faculty
cur.execute('''
CREATE TABLE IF NOT EXISTS Faculty (
    Faculty_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Email TEXT,
    Department TEXT
)
''')

# Create table for Librarians
cur.execute('''
CREATE TABLE IF NOT EXISTS Librarians (
    Librarian_ID INTEGER PRIMARY KEY,
    Name TEXT NOT NULL,
    Email TEXT
)
''')

# Create table for Transactions
cur.execute('''
CREATE TABLE IF NOT EXISTS Transactions (
    Transaction_ID INTEGER PRIMARY KEY,
    Book_ID INTEGER,
    Title TEXT,
    Student_Faculty_ID INTEGER,
    Librarian_ID INTEGER,
    Borrowed_Date DATE,
    Returned_Date DATE,
    FOREIGN KEY (Book_ID) REFERENCES Books (Book_ID),
    FOREIGN KEY (Title) REFERENCES Books (Title),
    FOREIGN KEY (Student_Faculty_ID) REFERENCES Students (Student_ID),
    FOREIGN KEY (Librarian_ID) REFERENCES Librarians (Librarian_ID)
)
''')

# Commit the changes
conn.commit()

# Close the database connection
conn.close()
```
